## Q.3 Report some descriptive statistics regarding this data include atleast one measure of central tendancy and atleast one measure of variablity?




```python
#Q.3 solution
import pandas as pd
import numpy as np
import math
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import t as pt
%matplotlib inline
#Read the data and print it
d=pd.read_csv('Untitled spreadsheet - Sheet1.csv')
print(d)
#Detail about the data set
print(d.describe())
#Find the mean median and standard deviation of both column of dataFrame
cong_mean=d['Congruent'].mean()
incong_mean=d['Incongruent'].mean()
cong_median=d['Congruent'].median()
incong_median=d['Incongruent'].median()
cong_std=d['Congruent'].std()
incong_std=d['Incongruent'].std()
a=[[int(24),int(24)],[cong_mean,incong_mean],[cong_median,incong_median],[cong_std,incong_std]]
data=pd.DataFrame(a,index=['Sample Size','Mean','Median','Standard Deviation'],columns=['Congruent','Incongruent'])
#printing the mean meadian and standard deviation of both column Congruent and Incongruent
print(data)
```

## Q.5 Now,perform the statistical test and report your result .What is your confidence level and your crictical statistic value ? Do you reject the null hypothesis or fail to reject it ? Come to  a conclusion in terms of experiment task .Did the results match up with your expectation ?
Ans
-  Sample Size(n)=24
-  Degree of freedom (df)=n-1=24-1=23
-  Test is two tailed t-test.Here we find crictical value of t at 95% confidence level

```python
#Q.5
#Finding the t-critical value for 95% confidence interval for two tailed test and here degree of 
#freedom is 23
print(pt.ppf(0.975,23),5)
#making an new column which contain the difference of this two column
d['difference']=d['Congruent']-d['Incongruent']
#finding the mean and standard deviation of difference dataset
diff_mean=d['difference'].mean()
diff_std=d['difference'].std()
print(round(diff_mean,5))
print(round(diff_std,5))
#finding the t-value
t_value=diff_mean/(diff_std/math.sqrt(24))
print(t_value)
```
